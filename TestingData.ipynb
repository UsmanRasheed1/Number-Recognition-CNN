```python
import cv2
import tensorflow as tf
import numpy as np
import os
import matplotlib.pyplot as plt

def prepare(filepath):
    IMG_SIZE = 100  
    img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE) 
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE)) 
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1) / 255.0  

def evaluate_model(model, directory):
    i = -1
    j = 0
    correct = 0
    for filename in os.listdir(directory):
        if filename.endswith(".jpg"):
            file_path = os.path.join(directory, filename)
            predictions = model.predict(prepare(file_path))
            predicted_class = np.argmax(predictions, axis=1)

            if(j % 3 == 0):
                i = i + 1
            j = j + 1
            if(i == predicted_class[0]):
                correct = correct + 1
            
    return (correct / j) * 100

# Load all three models
model_paths = ["MNIST-Project-CNN.keras", "MNIST-Project-CNN-ver2.keras", "MNIST-Project-CNN-ver3.keras"]
models = [tf.keras.models.load_model(path) for path in model_paths]

# Directories for test data
directories = ["D:/Keras/Written Numbers", "D:/Keras/MsPaint"]

# Evaluate all models on both datasets
results = []
for model in models:
    model_results = []
    for directory in directories:
        accuracy = evaluate_model(model, directory)
        model_results.append(accuracy)
    results.append(model_results)
    
trainingdata = [71.53, 90.52, 94.94]

for i in range(len(results)):
    results[i].insert(0, trainingdata[i])

# Save the results for the next cell
results
```

```python


# Plotting the results
labels = ['Training Data','Written Numbers', 'MsPaint']
x = np.arange(len(labels))

fig, ax = plt.subplots()
width = 0.2

for i in range(len(models)):
    ax.bar(x + i*width, results[i], width, label=f'Model Version {i+1}')


y_ticks = np.arange(0, 101, 10)
ax.set_yticks(y_ticks)
ax.set_ylabel('Accuracy (%)')
ax.set_title('Model Accuracy on Different Datasets')
ax.set_xticks(x + width / 2)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```
